```python
import sympy as sp
# question 4a

x, y = sp.symbols('x y')
f = sp.exp(x)* sp.sin(y) + y**3
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print ("first partial wrt x:", df_dx)
print ("first partial wrt y:", df_dy)
```

```python
import sympy as sp
# question 4b

x, y = sp.symbols('x y')
g = x**2*y+x*y**2

dg_dx = sp.diff(g, x)
dg_dy = sp.diff(g, y)

dg_dx_point = dg_dx.subs({x: 1, y: -1})
dg_dy_point = dg_dy.subs({x: 1, y: -1})

gradient_vector = [dg_dx_point, dg_dy_point]

magnitude_at_point = sp.sqrt(dg_dx_point**2 + dg_dy_point**2)

print("gradient vector:", gradient_vector)
print("magnitude of the gradient vector at (1, -1):", magnitude_at_point)
```

```python
import sympy as sp
# question 4c

x, y = sp.symbols('x y')
h = sp.ln(x**2+y**2)

dh_dx = sp.diff(h, x)
dh_dy = sp.diff(h, y)

d2h_dx2 = sp.diff(dh_dx, x)
d2h_dy2 = sp.diff(dh_dy, y)
d2h_dxy = sp.diff(dh_dx, y)
d2h_dyx = sp.diff(dh_dy, x)

"""
The mixed partials are symmetric. This means that if you were to interchange the variables, you would get the same function. Since they
are the same, they are also continuous. This suggests that the function is smooth and have the same rate of change, no matter the direction
being approached from. Second partials are often equivalent
"""

print ("first partial wrt x:", dh_dx)
print ("first partial wrt y:", dh_dy)
print ("second partial wrt xx:", d2h_dx2)
print ("second partial wrt yy:", d2h_dy2)
print ("second partial wrt xy:", d2h_dxy)
print ("second partial wrt yx:", d2h_dyx)
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
# question 4e

x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
# question 4e

x, y = sp.symbols('x y')
r = 2*x**2 - 3*y**2 
r_func = sp.lambdify((x, y), r, 'numpy')

x_vals = np.linspace(-2, 2, 500)
y_vals = np.linspace(-2, 2, 500)
X, Y = np.meshgrid(x_vals, y_vals)
Z = r_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $r(x, y) = 2x^2 - 3y^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
